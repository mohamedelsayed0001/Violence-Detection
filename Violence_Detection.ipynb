## Violence Detection with YOLO V11 
This project leverages the YOLOv11 small model to detect violent activities in real-time video streams.

The model has been fine-tuned using `the New Violence` Dataset, sourced from `Roboflow`.

```python
# Importing necessary libraries
from ultralytics import YOLO
import cv2
```

```python
# Load a model
model = YOLO("yolo11s.pt")
```

---
### Training


```python
# Training the YOLO model
train_results = model.train(
    data="new_violence_dataset.v3i.yolov11/data.yaml",   
    epochs=10, 
    imgsz=640, 
    device="cpu",  
)
```

---
### Testing the Model on Live Video

```python

def predict_frame_yolo(frame, model):
    """
    Pass the frame to a YOLO model and get predictions (bounding boxes and labels).
    """
    # Perform inference
    results = model(frame)  # YOLO model inference
    return results[0].boxes  # Return bounding boxes, confidence scores, and class IDs

def draw_predictions(frame, predictions, class_names, conf_threshold=0.5):
    """
    Draw bounding boxes and labels on the frame based on YOLO predictions.
    """
    for box in predictions:
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Get bounding box coordinates
        conf = box.conf.item()  # Get confidence score
        class_id = int(box.cls.item())  # Get class ID

        # Filter by confidence threshold
        if conf >= conf_threshold:
            label = f"{class_names[class_id]} ({conf:.2f})"
            color = (0, 255, 0)  # Green for bounding boxes

            # Draw bounding box
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), color, 2)

            # Draw label
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return frame

def process_video_yolo(input_path, output_path, model, class_names):
    """
    Process video frame by frame, make predictions using a YOLO model, overlay predictions, and save output.
    """
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print("Error: Cannot open video.")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Initialize video writer
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Predict using the YOLO model
        predictions = predict_frame_yolo(frame, model)

        # Overlay predictions on the frame
        frame = draw_predictions(frame, predictions, class_names)

        # Write the frame to the output video
        out.write(frame)

    # Release resources
    cap.release()
    out.release()
    print("Video processing completed and saved to", output_path)

# Usage
input_video_path = "Fighting028_x264.mp4"
output_video_path = r"output.mp4"  

# Class names for the YOLO model
class_names = model.names

# Process the video
process_video_yolo(input_video_path, output_video_path, model, class_names)
```
